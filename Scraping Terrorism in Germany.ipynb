# # function for logic

```python
def wiki_table():
    thead = [i.text.strip() for i in table2.find_all('th')]
    data=[]
    for r in soup.select('table.wikitable.sortable.collapsible.mw-collapsible.mw-made-collapsible.jquery-tablesorter')[0].tbody.find_all('tr'):
        data.append([i.text.strip() for i in r.find_all('td')])
    df = pd.DataFrame(data)
    df.columns = thead
    return df
```

```python
wiki_table()
```
